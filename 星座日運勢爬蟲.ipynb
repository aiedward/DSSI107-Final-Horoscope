```python
from bs4 import BeautifulSoup
import requests
import time
import csv
```

```python
def writefile(filename,inputlist):
    with open(filename,'a+', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile, delimiter=',')
        for row in inputlist:
            writer.writerow(row)
def readfile(filename,encode = 'UTF-8'):
    rawlist = []
    with open(filename,'r+', encoding=encode) as csvfile:
        rows = csv.reader(csvfile)
        for row in rows:
            rawlist.append(row)
    return rawlist
```

# 因應網站結構，找我們需要的日運勢

```python
urllist = []
#for i in range(64,363):
#for i in range(66,411):
for i in range(64,324):
    try:
        urllist.append([a_tags[i].text,a_tags[i].attrs['href']])
    except:
        print(i,'No URL')
print(len(urllist))
```

# 初步截取需要的內容，還不是很準確
### 先觀察文字內容有沒有固定，取頭(x)尾(y)，tmp = tmp[tmp.find('x')+length of x:tmp.find('y')]

```python
for i in range(0,len(urllist)):
    if (i%10==9):
        time.sleep(0.01)
        #print(i)
    horoscope_day = requests.get(urllist[i][1])
    if horoscope_day.status_code == requests.codes.ok:
        soup = BeautifulSoup(horoscope_day.text, 'html.parser')
    tmp = soup.get_text()
    if(urllist[i][0].find('唐綺陽')):
        tmp = tmp[tmp.find('歡迎按讚加入fb分享星粉絲團，接收最新星座運')+23:tmp.find('延伸閱讀')]
    else:
        tmp = tmp[tmp.find('歡迎按讚加入fb分享星粉絲團，接收最新星座運勢喔')+26:tmp.find('延伸閱讀')]
    urllist[i].append(tmp)
```

# 寫檔

```python
writefile('DSSFinal_daily_horoscope.csv',urllist)
```
